### Création de mask et mise à l'échelle d'images

```python
#Import de bibliothèques

import numpy as np
import os
import skimage as ski
import matplotlib.pyplot as plt
from skimage.transform import resize
from skimage import io, img_as_ubyte
```

```python
#Chargement image
image = ski.io.imread("images/test1.jpg")
```

```python
#Convertion des images en (256,256,3)

resize_image = resize(image, (256,256,3), anti_aliasing=True)
```

```python
#resize_image = image
```

```python
#affichage du resulat

fig, axes = plt.subplots(1, 2, figsize=(10, 5))
ax = axes.ravel()

ax[0].imshow(image)
ax[0].set_title("image originale")

ax[1].imshow(resize_image)
ax[1].set_title("image en 256x256")

plt.tight_layout()
plt.show()
```

```python
#Création de différents masks
"""
mask = np.ones((256,256,3), dtype= "uint8")
mask[:,128:138,:] = 0
#verticale_image_masked = np.multiply(mask, resize_image)
resize_image[:,128:138,:] = 0.35*np.ones((256,10,3), dtype= "uint8")
mask *= 255"""
```

```python
mask = np.ones((256, 256, 3), dtype= "uint8")
for make_black in range(0, 240, 10):
    mask[:, make_black: make_black+4, :] = 0
    try: 
        resize_image[:, make_black: make_black+4, :] = 0.35*np.ones((256, 4, 3), dtype= "uint8")
    except ValueError:
        resize_image[:, make_black: make_black+4, :] = 0.35*np.ones((256, 4, 3), dtype= "uint8")
#verticale_image_masked = np.multiply(mask, resize_image)
mask *= 255
```

```python
#affichage du resulat

fig, axes = plt.subplots(1, 2, figsize=(10, 5))
ax = axes.ravel()

ax[0].imshow(resize_image)
ax[0].set_title("image masked")

ax[1].imshow(mask)
ax[1].set_title("mask")

plt.tight_layout()
plt.show()
```

```python
#Convertion en uint8
verticale_image_masked = img_as_ubyte(resize_image)
mask = img_as_ubyte(mask)
```

```python
ski.io.imsave("000001.png",verticale_image_masked)
ski.io.imsave("000001_m.png",mask)
```
